```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *


# Define a function to perform the experiments for different gamma values
def analyze_iterations(step_sizes, gamma_values, x0):
    plt.figure(figsize=(14, 7))
    for gamma in gamma_values:
        # Construct A matrix for the current gamma
        L = np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta), np.cos(theta)]])
        D = np.array([[gamma, 0], [0, 2 - gamma]])
        A = L @ D @ np.linalg.inv(L)

        num_iters_list = []

        for step_size in step_sizes:
            _, _, num_iters = gradient_descent(A, x0, step_size)
            num_iters_list.append(num_iters)

        # Plot number of iterations vs step size
        plt.plot(step_sizes, num_iters_list, marker='o', label=f' = {gamma}')

    plt.xlabel('Step size')
    plt.ylabel('Number of iterations')
    plt.title('Number of Iterations vs Step Size')
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.grid(True)
    plt.show()

# Define parameters
theta = np.pi / 4
step_sizes = [0.5, 0.4, 0.3, 0.2, 0.1, 0.05, 0.02]  # Extended range of step sizes
x0 = np.array([2, 0], dtype=float)

# Perform the analysis for different gamma values
analyze_iterations(step_sizes,gamma_values=[1/4, 1/16], x0=x0)
```
