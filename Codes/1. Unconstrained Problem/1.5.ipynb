```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *

# Define parameters
theta = np.pi / 4
x0 = np.array([2, 0], dtype=float)

# Gamma values
gamma_values = [1, 1/2, 1/4, 1/8, 1/16]

# Arrays to store results
exact_iters = []
backtracking_iters = []

# Perform analysis for each gamma value
for gamma in gamma_values:
    L = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    D = np.array([[gamma, 0], [0, 2 - gamma]])
    A = L @ D @ np.linalg.inv(L)
    
    # Gradient descent with exact line search
    exact_num_iters = gradient_descent_exact(A, x0)
    exact_iters.append(exact_num_iters)
    
    # Gradient descent with backtracking line search
    backtracking_num_iters = gradient_descent_backtracking(A, x0)
    backtracking_iters.append(backtracking_num_iters)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(gamma_values, exact_iters, marker='o', label='Exact Line Search')
plt.plot(gamma_values, backtracking_iters, marker='o', label='Backtracking Line Search')
plt.xlabel('Gamma (γ)')
plt.ylabel('Number of Iterations')
plt.title('Number of Iterations vs Gamma for Different Line Search Methods')
plt.xscale('log')  # Log scale for gamma values
plt.legend()
plt.grid(True)
plt.show()
```
